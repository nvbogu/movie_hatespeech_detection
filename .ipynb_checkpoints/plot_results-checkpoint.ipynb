```python
import pandas as pd
import nltk
```

```python
df_result = pd.read_csv("data/results.csv", index_col=0)
df_result.head(1)
```

```python
def replace_val(df):
    df = df.replace('neither',0)
    df = df.replace('not hate speech',0)
    df = df.replace('offensive language',1)
    df = df.replace('hate speech',2)
    return df
```

```python
def calculate_keyword_formular(count):
    if count == 1:
        return 1
    elif count == 2:
        return 2
    elif count==0:
        return 0
    else:
        return 3
```

```python
df_result['we_label_fox_news'] = replace_val(df_result['we_label_fox_news'])
df_result['we_label_twitter'] = replace_val(df_result['we_label_twitter'])
df_result['bert_label_twitter'] = replace_val(df_result['bert_label_twitter'])
df_result['bert_label_fox_news'] = replace_val(df_result['bert_label_fox_news'])
df_result['lstm_label_twitter'] = replace_val(df_result['lstm_label_twitter'])
df_result['lstm_label_fox_news'] = replace_val(df_result['lstm_label_fox_news'])
```

```python
df_result['keyword_count_formular'] = df_result['hate_keyword_count_unique'].apply(calculate_keyword_formular)
```

```python
df_result.head(1)
```

```python
df_result['hate_score'] = (df_result['we_label_fox_news']
                                    +df_result['we_label_twitter']
                                    +df_result['bert_label_twitter']
                                    +df_result['bert_label_fox_news']
                                    +df_result['lstm_label_twitter']
                                    +df_result['lstm_label_fox_news']
                                    +df_result['keyword_count_formular'])
```

```python
df_result['hate_score_normlized'] = round(df_result['hate_score']/df_result['hate_score'].max(),2)
```

## add bins

```python
df_result['ensemble_label'] = pd.cut(df_result['hate_score_normlized'], bins=[0, .5, .8, 1], labels=['neither', 'offensive', 'hate'])
```

# Analysis

# Creating Movie Metadata Dataframe

```python
mean_hate_score_by_movie_df = df_result.groupby('movie_name').hate_score_normlized.mean().to_frame('mean_hate_score').reset_index().sort_values(by='mean_hate_score', ascending=False)
```

```python
mean_hate_score_by_movie_df.head(1)
```

```python
new_label_distribution_df = df_result.groupby(['movie_name', 'tag']).ensemble_label.value_counts().to_frame('count_bin').reset_index()
new_label_distribution_df.head(3)
```

```python
new_label_distribution_df[new_label_distribution_df.movie_name=='friendship']
```

```python
df = new_label_distribution_df.groupby(['tag', 'ensemble_label']).count_bin.sum().to_frame('count').reset_index()
df_pivot = pd.pivot_table(
    df,
    values="count",
    index="tag",
    columns='ensemble_label'
)
```

```python
ax = df_pivot.plot(kind='bar', title='Ensemble, Label by tag distribution', color=['C1', 'C2', 'C0'])
```

```python
hate_df = new_label_distribution_df[new_label_distribution_df.ensemble_label=='hate'].groupby('tag').count_bin.mean() .to_frame('hate')
```

```python
offensive_df = new_label_distribution_df[new_label_distribution_df.ensemble_label=='offensive'].groupby('tag').count_bin.mean().to_frame('offensive')
```

```python
neither_df = new_label_distribution_df[new_label_distribution_df.ensemble_label=='neither'].groupby('tag').count_bin.mean().to_frame('neither')
```

```python
all_df = pd.concat([hate_df, offensive_df, neither_df], axis=1)
```
